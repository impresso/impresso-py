```python
from impresso import connect

impresso = connect()
```

# Search topics

Find topics in the Impresso database. Topics are generated using topic modeling techniques and represent themes or subjects discussed in the content items.

```python
impresso.topics.find()
```

Search for topics by keyword. This searches within the top words associated with each topic.

```python
impresso.topics.find(term="chat")
```

## Order results

Topics can be ordered by different fields.

```python
impresso.topics.find(term="fait", order_by="-name")
```

## Pagination

Iterate through all pages of results.

```python
result = impresso.topics.find(
    term="fait",
    limit=10,
)

print(f"Total topics in the result set: {result.total}")
for page in result.pages():
    print(
        f"Got page {page.offset} - {page.offset + page.size} of {page.total}. "
        + f"The first topic has {page.pydantic.data} content items"
    )
```

## Get a topic by its ID

Retrieve a specific topic using its unique identifier.

```python
# First, let's get a topic ID from the search results
result = impresso.topics.find(limit=1)
topic_id = result.pydantic.data[0]['id']
print(f"Topic ID: {topic_id}")

# Now get the full topic details
impresso.topics.get(topic_id)
```

## View as DataFrame

Convert the results to a pandas DataFrame for easier analysis.

```python
result = impresso.topics.find(limit=20)
result.df
```

## Explore topic words

Each topic contains a list of the most representative words with their probabilities.

```python
result = impresso.topics.find(limit=1)
topic = result.raw['data'][0]

print(f"Topic ID: {topic['id']}")
print(f"Language: {topic['language']}")
print(f"Content items count: {topic.get('contentItemsCount', 'N/A')}")
print(f"\nTop words:")
for word in topic.get('words', [])[:10]:
    print(f"  {word['w']}: {word['p']:.4f}")
```
