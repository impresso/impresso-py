```python
from impresso import connect

impresso = connect()
```

## Search images

```python
result = impresso.images.find(term="rocket")
result
```

## Image embedding search

```python
image_id = "luxwort-1930-09-26-a-i0036"
image = impresso.images.get(image_id)
image
```

```python
embeddings = impresso.images.get_embeddings(image_id)
embeddings
```

```python
impresso.images.find(
  embedding=embeddings[0],
  limit=3
)
```

```python
impresso.images.find(
  embedding=embeddings[1],
  limit=3
)
```

### Out of corpus embeddings

```python
impresso.images.find(
  embedding=impresso.tools.embed_text(text="belval", target="image"),
  # media_id="luxwort",
  limit=3
)
```

```python
impresso.images.find(
  embedding=impresso.tools.embed_image(image="https://impresso-project.ch/assets/images/posts/radio_and_newspapers_conference_2026.jpg", target="image"),
  limit=3
)
```
