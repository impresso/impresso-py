```python
result = impresso.images.find(term="rocket",content_type="object")
result
```

## Image embedding search

```python
image_id = "luxwort-1930-09-26-a-i0036"
image = impresso.images.get(image_id, include_embeddings=True)
image
```

```python
embeddings = impresso.images.get_embeddings(image_id)
embeddings
```

```python
impresso.images.find(
  embedding=embeddings[0],
  limit=3
)
```

```python
impresso.images.find(
  embedding=embeddings[1],
  limit=3
)
```

## Get content items related to images

```python
result = impresso.images.find(term="rocket",content_type="object")
result
```

```python
issue_ids = result.df['issueId'].unique().tolist()
issue_ids
```

```python
from impresso import OR

content_items = impresso.search.find(
  issue_id=OR(*issue_ids)
)
content_items
```
